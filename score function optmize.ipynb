```python
def draw(y1,y2,Tday=20):
    # 天数 Tday  
    Mt = np.random.rand(Tday)*(y2-y1) +y1   
    x = np.arange(y1,y2,0.1)
    y = np.zeros_like(x)
    for i in range(len(x)):
        y[i] = np.sum(np.fabs((x[i]-Mt)/(x[i]+Mt)))/Tday
    plt.figure(figsize=(14,7))
    plt.plot(x,y)
    plt.title('Mt range from '+ str(y1)+ ' to '+ str(y2)+';   Tday = '+str(Tday) )
```

```python
draw(50,260,100)
```
